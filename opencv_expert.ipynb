```python
import os
import json
from crewai import Agent, Task, Crew, Process
from crewai_tools import JSONSearchTool
from langchain_openai import ChatOpenAI
from textwrap import dedent
os.environ["OPENAI_API_KEY"] = ""

search_tool = JSONSearchTool()

# Define your agents with roles and goals
researcher = Agent(
  role='RAG expert from JSON',
  goal='Solve user doubts based on the knoledge base of /content/opencv.json file',
  backstory="""You work at a OPENCV.
  Your expertise lies in finding right answers to the questions asked by user using the json file.
  You have a knack for extracting data from json file.""",
  verbose=True,
  allow_delegation=False,
  tools=[search_tool],
  llm=ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
)



#create task
def RAG_TASK(question):
		return Task(description=dedent(f"""\
			User query: {question}.
			Take input from user and try to answer user query in appropriate and correct manner
			"""),
			expected_output="correct answer and guide which specific OPENCV functions to use based on usecase ",
      agent=researcher
		)



question = input(
    dedent("""
      What do you want to know about Opencv?
    """))
task1 = RAG_TASK(question)

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[researcher],
  tasks=[task1],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)
# Get your crew to work!
result = crew.kickoff()

print("######################")
print(result)
```
